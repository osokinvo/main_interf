```python
def interf(lambda1, r1, r2, d, n=1.7):
    if r2 > 1 - r1:
        r2 = 1 - r1
    return r1 + r2*np.cos(4*np.pi*n*d/lambda1)
```

```python
lambda_blue = 464.2
lambda_green = 549.1
lambda_red = 611.4
```

<h2>R1 60% R2 30%<h2>

```python
r1 = 0.6
r2 = 0.3
```

```python
d_max = 1500
img_1050 = [[interf(lambda_blue, r1, r2, d), interf(lambda_green, r1, r2, d), interf(lambda_red, r1, r2, d) ] \
    for d in range(d_max)]
img = np.array(img_1050)

# cv.imshow('Disp', img)
# k = cv.waitKey(0)
```
